```python
import pandas as pd
import matplotlib.pyplot as pit
url='/content/Test.csv'
df = pd.read_csv(url)
df
```

```python
mean_value = df['Item_MRP'].mean()
print(f"Mean : {mean_value}")
median_value = df['Item_MRP'].median()
print(f"Median : {median_value}")
mode_value = df['Item_MRP'].mode()
print(f"Mode : {mode_value}")
std_value = df['Item_MRP'].std()
print(f"Standard deviation: {std_value}")
var_value = df['Item_MRP'].var()
print(f"Variace : {var_value}")
kurtosis_value = df['Item_MRP'].kurtosis()
print(f"kurtosis : {kurtosis_value}")
skew_value = df['Item_MRP'].skew()
print(f"skewness : {skew_value}")

```

```python
df['Item_MRP'].value_counts().plot(kind='pie')
pit.xlabel('Item_MRP')
pit.ylabel('Count')
pit.title('Item_MRP Distribution')
pit.show()
```
